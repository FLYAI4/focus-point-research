```python
import requests
from PIL import Image
from io import BytesIO
from diffusers import StableDiffusionUpscalePipeline
import torch

# load model and scheduler
model_id = "stabilityai/stable-diffusion-x4-upscaler"
pipeline = StableDiffusionUpscalePipeline.from_pretrained(
    model_id, revision="fp16", torch_dtype=torch.float16
)
pipeline = pipeline.to("cuda")
```

```python
url = "https://cdn.veritas-a.com/news/photo/old/4/3_1165480082.jpg"
response = requests.get(url)
low_res_img = Image.open(BytesIO(response.content)).convert("RGB")

# 원래 비율로 크기 조정
width, height = low_res_img.size
max_size = max(width, height)
new_width = int(width * 128 / max_size)
new_height = int(height * 128 / max_size)
resized_img = low_res_img.resize((new_width, new_height))

prompt = "The Starry Night"

upscaled_image = pipeline(prompt=prompt, image=resized_img).images[0]
upscaled_image.save("Starry Night.png")
```
